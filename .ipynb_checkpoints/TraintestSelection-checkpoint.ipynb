```python
import os,glob,cv2
```

```python
CurrentDirectory=os.getcwd()
TestImagePath=CurrentDirectory+'/FINAL/TrainImage/'
```

```python
test = glob.glob(TestImagePath+"*.png")
test.sort()
print(test)
```

```python
path=CurrentDirectory+'/MyGT/'
pathsave=CurrentDirectory+'/FINAL/TrainGT/'

teststore=[]

for imname in test:
    fName, ext = os.path.splitext(imname)
    name=fName[-3:]
#     print(name)
    teststore.append(name)
    GTName=path+name+ext
    print(GTName)
    Gt=cv2.imread(GTName,0)
    savePath=pathsave+name+ext
    cv2.imwrite(savePath,Gt)    
```
