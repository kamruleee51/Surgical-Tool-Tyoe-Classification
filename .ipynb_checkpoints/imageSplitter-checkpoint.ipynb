```python
import os,glob,cv2
import numpy as np
```

```python
CurrentDirectory=os.getcwd()

allImages=CurrentDirectory+'/MyOR/'

testPath=CurrentDirectory+'/FINAL/TestImage/'
trainPath=CurrentDirectory+'/FINAL/TrainImage/'

imageOR = glob.glob(allImages+"*.png")
imageOR.sort()

# print(imageOR)

import random
test=[]
for i in range(220):
    r=random.randint(1,len(imageOR))
    if r not in test:
        test.append(r)
        
test.sort()
print(test)
print(len(test))

totalNumber=np.arange(len(imageOR))

train=[v for v in totalNumber if v not in test]

train.sort()
print(train)
print(len(train))
```

```python
k=0
for j in range(len(test)):
    testindex=test[j]
    testimg=cv2.imread(imageOR[testindex],1)
    outName=imageOR[testindex].replace( allImages, testPath) 
    cv2.imwrite(outName, testimg)
    k=k+1
    print(k)
    print(imageOR[testindex])
```

```python
k=0
for j in range(len(train)):
    testindex=train[j]
    testimg=cv2.imread(imageOR[testindex],1)
    outName=imageOR[testindex].replace( allImages, trainPath) 
    cv2.imwrite(outName, testimg)
    k=k+1
    print(k)
    print(imageOR[testindex])
```
